## 과제 2번

- 과제 2번은 Carousel Component를 통해 자동으로 재생되는 이미지를 만들고, 그 아래 화면의 크기에 따라서 각 행에 표시되는 카드의 갯수를 다르게 하는 것이었다.

- 어려웠던 점 : 대부분의 이미지 자동재생등은 bootstrap에서 코드를 그대로 가져와서 힘들지는 않았지만 화면의 크기에 따라서 표시하는 것이 어려웠다.
  화면의 크기에 따라서 표시하는 부분에서 각 요소들 사이에 여백을 줘야하는데 이를 위해 gutter를 사용했지만 원하는 대로 구현되지 않았다.
      <div class="col-12 col-md-6 col-lg-4">
      <div class="card" style="height: 100%;">
        <img src="images/movie6.jpg" class="card-img-top" alt="movie6" style="height: 70%;">
        <div class="card-body">
          <h5 class="card-titl">Moive title</h5>
          <p class="card-text">This is a longer card with supporting text below as a natural lead-in to additional content. This content is a little bit longer</p>
        </div>
        </div>
    </div>
    이 부분에서 class card부분에 col을 주면 카드 영역에서 안쪽에서 padding이 형성되서 여백이 형성되지 않는다. 하지만, div를 밖에 주고 col을 주면 card가 형성되기 전에 영역에 대해 padding이 형성되므로 여백을 원하는대로 줄 수 있었다.
