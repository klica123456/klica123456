**Machine Learning**
- 기계 학습 : 컴퓨터가 데이터로부터 새로운 규칙/수식을 도출해내는 직업
- 기계학습의 실무 핵심 Point
    - 학습 능력 : 컴퓨터가 기존의 데이터를 얼마나 잘 학습하는가?
    - 일반화 능력 : 새로운 데이터가 들어왔을 때, 학습한 수식(Model)이 얼마나 잘 예측하는가?
- 기계학습의 핵심 3요소
    - 데이터(교과서) : 학습 목적에 맞는 깔끔한 데이터 셋 구축 - 데이터 클리닝
                        특성 공학 (Feature Engineering) 데이터를 깔끔하게 다듬는 과목
    - 알고리즘-수학적인 구조 (선생님) : 학습 목적에 맞는 적절한 알고리즘을 선택 
                                        선형회귀분석 / 의사결정나무 / KNN / SVM / Ensenble
    - 하드웨어 (학생) : CPU / GPU
                           -> 비용 (Cost)
**Machine Learning 종류**
- 지도 학습 (Supervised Learning) : 목표변수(Y)와 설명변수(X)간의 관계를 수식화하여,
                                    새로운 설명변수(X)에 대해 목표변수(Y)를 예측하거나 분류하는 법
    - 회귀 (Regression, Y-연속형) : 정확한 목표변수를 예측
    - 분류 (Classification, Y-범주형) : 특정 항목을 정확하게 구분
    EX) 주가 예측 프로그램 -> Y주가 (회귀) / 스팸 메시지 분류기 -> Y스팸여부 (분류)
    
    - 지도 학습 절차
        - 데이터 핸들링 (데이터 불러오기, 파생변수(기존의 변수를 조합하여 새로운 변수를 만들어 내는 것) 생성, 이상치 제거...)
        - 회귀/분류 기법을 적용시킬 목표변수(Y)와 설명변수(X)를 설정
        - 학습 데이터(Train Set)와 검증 데이터(Test Set)를 분할
        - 학습데이터를 이용해, 수식을 생성(Modeling, 학습)
            - 특성 공학 (Feature Engineering)
            - 학습 알고리즘
        - 평가 (Evaluation)
            - 학습 능력 -> 학습 성능 평가(Train Set)
                - 처음에 X_train과 Y_train으로 model을 만들고 이 model에 X_train을 다시 집어넣어 얼마나 잘 학습했는지 Y_train_pred가 나오는데 얘를 Y_train과 비교
                    Y_train(실제값)-Y_train_pred(예측값)
            - 일반화 능력 -> 일반화 성능 평가(Test Set)
                - X_test를 model에 집어넣어서 Y_test_pred를 얻어 정답인 Y_test와 비교
                    Y_test(실제값) - Y_test_pred(예측값)
- 비지도 학습 (Unsupervised Learning) : 설명변수(X)들 간의 관계/연관성등을 파악해 비슷한 데이터끼리 **군집화** 하거나 연관성 있는 데이터를 묶는 기법
    - 군집분석 (Clustering) : 서로 유사한, 특성이 비슷한 데이터끼리 묶어주는 작업
    - 연관분석 (Association Analysis) : 데이터간 유사도를 계산하여, 서로 연관성 높은 데이터를 찾는 기법
    Ex) 장바구니 분석 / 추천 시스템 ... 

- 강화 학습 (Reinforcement Learning) : 컴퓨터 시뮬레이션을 통해 주어진 환경에 대해 사용자가 주는 보상이 좋은 방향으로 학습하는 기법 - 데이터를 스스로 만들어 냄
    -> 데이터가 없이도 학습이 가능 but.어려움

```python
# Wafer 폐기여부를 판별하는 모델을 생성
# 불량갯수 알고, 폐기여부 앎
df1['target_binom'].value_counts()
```

```python
df1.columns
```

```python
Y = df1['target_binom']
# drop() : 데이터 프레임에서 불필요한 항목을 제거
X = df1.drop(columns=['Target', 'Error_message', 'target_binom', 'Chamber_Route',
                     'Unnamed: 0', 'Ox_Chamber', 'process', 'type', 'No_Die',
                     'Unnamed: 0_x', 'photo_soft_Chamber', 'process 2',
                    'Unnamed: 0_y', 'Etching_Chamber', 'Process 3',  
                     'Chamber_Num', 'process4'])
# 문자데이터들도 학습에 참여시킬 수 있음
```

```python
# 숫자데이터만 선택
numerlic_list = X.describe().columns.tolist() # 숫자인 데이터들만 통계량 내서 columns로 바꾸고 list로 만듦
X1 = X[numerlic_list]
```

```python
pip install scikit-learn
```

```python
# 학습데이터와 검증데이터 분할
from sklearn.model_selection import train_test_split
```

```python
# 학습 데이터와 검증데이터를 8:2 비율로 분할
# 데이터마다 비율이 다를 수도 있음
X_train, X_test, Y_train, Y_test = train_test_split(X1, Y, random_state=1234) # 랜덤인데 1234가 같은면 같은 데이터로 나뉨
# random_state : 무작위 추출을 원하는 index로 맞추어 추출
print(X_train.shape) # shape : 데이터의 행과 열개수를 출력
print(X_test.shape)
```

```python
# 학습 수행
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)# fit : 학습을 수행하는 함수
# 학습의 결과
```

```python
# 결측값 (Missing Value)
# 1) 제거 : 결측값이 있는 행을 삭제
# 2) 대치 : 다른 값을 결측값에 대채 (평균 / 중앙값) -> 알고리즘 (knn)
```

```python
model
```

```python
#평가
from sklearn.metrics import accuracy_score
```

```python
# 정확도 : 전체 데이터 중 정확하게 분류한 데이터의 수 0~1
Y_train_pred = model.predict(X_train) # predict() model에 X를 집어넣을때 결과를 뽑아주는 함수
accuracy_score(Y_train, Y_train_pred)
# 1.0이 정확도 100%
# 학습능력
```

```python
#일반화 능력
Y_test_pred = model.predict(X_test)
accuracy_score(Y_test, Y_test_pred)
```

- 분류 모델에서 평가 방법(Evaluation)

- 정확도 = 정확하게 분류한 데이터 수 / 전체 데이터 수

- 오차 행렬 (Confusion Matrix) : 나중에 예측 model만들때 positive를 기준으로 주로 계산되므로 보통 불량을 판별할려고 model을 형성하므로 positive를 불량으로 놓음

![nn](캡처.png)

- 정밀도 (Precision) = TP/ (TP+FP) 불량으로 예측한 것 중 실제 불량의 비율
- 재현율 (Recall) = TP / (TP+FN) 실제 불량 중 불량을 불량으로 예측한 비율
- F1 Score = 정밀도 + 재현율로 2*정밀도*재현율/(정밀도+재현율)

Real Negative, Real Positive, Predict Negative, Predict Positive 로 나눔

EX) 공장의 조업조건(X)을 이용해 품질을 분류하는 모델을 생성
Y(불량여부) 1000개 / 정상 950 / 불량 50
다 정상으로 예측하는 model이 나올 수도 있음 그래
서 정확도만 가지고 판단하면 안됨
-> Model : 950/1000 = 95%
-> Model 정밀도 : 10/20 = 10% - 정상을 불량으로 판단한 것은 큰 문제가 되진 않지만 불량을 정상으로 판단하면 고객한테 가므로 문제가 됨 따라서, 보통 불량을 기준으로 판단함
-> Model 재현율 : 10/50 = 20%
